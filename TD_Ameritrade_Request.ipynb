# Read in Data

```python
from config import API_KEY
import pandas as pd
import requests
```

The list of NASDAQ and NYSE stock tickers were sourced from http://eoddata.com/stocklist/NASDAQ/A.htm?e=NASDAQ&l=A. 

```python
stock_symbols = pd.read_excel(r'C:\Users\ASUS\Thinkful_Projects\Stock_Clustering\Stock_List.xlsx')
#stock_symbols
```

```python
n = 100
symbol_segments = [stock_symbols[i:i+n] for i in range(0, stock_symbols.shape[0], n)]
#symbols_chunk[0]
```

```python
# define the endpoint
endpoint = r"https://api.tdameritrade.com/v1/instruments"

stock_dict = {}

for segment in symbol_segments:
    payload = {
        'apikey': API_KEY,
        'projection': 'fundamental',
        'symbol': list(segment['Symbol'])
    }

    content = requests.get(url =  endpoint, params = payload)
    
    data = content.json()
    
    try:
        for stock in data:
            stock_dict[data[stock]['cusip']] = data[stock]['fundamental']
    except: 
        continue
stock_df = pd.DataFrame(stock_dict)
```

```python
stocks = stock_df.copy().T
stocks
```
